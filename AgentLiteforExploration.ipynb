# Financial Data Analysis with Enhanced AgentLite Framework

## Project Overview

This project extends the AgentLite framework to process and analyze various types of data, including financial data, stock market data, e-commerce data, and financial fraud data. The framework integrates a Large Language Model (LLM) like GPT-4 to generate insights, perform natural language processing tasks, and provide valuable context. The system features an interactive user interface built with Streamlit, allowing users to upload their data, review processed outputs, visualize insights, and interact with the data in real-time. The framework is designed to be modular and scalable, making it adaptable to various data-centric tasks.

## Features

- **Modular Data Processing**: Handles various datasets with customizable data processing pipelines via the `DataProcessingAgent`. It includes data cleaning, feature engineering, and transformation capabilities tailored to the specific data type (e.g., financial transactions, stock prices, e-commerce data).

- **LLM Integration**: Utilizes GPT-4 and the `promptgen` component to generate insights, summaries, and perform other NLP tasks based on the processed data. The integration allows for sophisticated, context-aware prompt generation and LLM interaction.

- **Interactive UI**: Built using Streamlit, the UI allows users to upload data files, interact with processed data, generate insights, and visualize results within an intuitive and user-friendly environment.

- **File Handling**: Supports ZIP and CSV file formats, automatically extracting and processing the data using the `FileHandler` module.

- **Advanced Data Visualization**: Visualizes processed data and LLM-generated insights using the `VisualizationAgent`, enabling users to quickly identify trends, correlations, and anomalies in the data.

- **Workflow Coordination**: The `AgentManager` orchestrates the entire workflow, ensuring seamless integration and execution of tasks by different agents, and adapts to the specific needs of the uploaded data.

## Use Cases

### **1. Stock Market Analysis**
- **Data Type**: Historical stock market data (e.g., Open, High, Low, Close prices, and Volume).
- **Functionality**: The framework processes historical stock data, generates features like moving averages, normalizes price data, and predicts future trends. The LLM provides market insights based on the data, and the Visualization Agent plots stock trends, correlations, and more.

### **2. Financial Fraud Detection**
- **Data Type**: Transactional data from financial institutions.
- **Functionality**: The framework cleans the data, creates features related to transaction patterns, and applies anomaly detection methods. The LLM generates risk assessments and insights on potential fraudulent activities. The Visualization Agent produces heatmaps and other visualizations to identify fraud patterns.

### **3. E-Commerce Data Analysis**
- **Data Type**: Customer reviews, product sales data, and transaction records.
- **Functionality**: The framework processes e-commerce data, segments customers, analyzes product performance, and performs sentiment analysis on reviews. The LLM generates summaries and customer insights, while the Visualization Agent creates customer segmentations and sentiment distribution charts.

## How It Works

### **1. User Interaction (Streamlit UI)**
- **File Upload**: Users interact with the Streamlit UI to upload a ZIP or CSV file containing their data. The data is initially displayed for review.
- **Data Display**: The uploaded data is shown in a tabular format within the Streamlit interface, allowing users to verify the data before processing begins.

### **2. File Handling**
- **File Processing**: The `FileHandler` processes the uploaded file, extracting data if it's a ZIP file or reading it directly if it's a CSV. The data is then loaded into a pandas DataFrame, ready for processing.

### **3. Data Processing**
- **Cleaning and Engineering**: The `DataProcessingAgent` cleans the data by handling missing values, performs feature engineering to create new features (e.g., moving averages, daily returns), and applies necessary transformations to prepare the data for analysis.
- **Transformation**: The data is further transformed (e.g., normalization, aggregation) to be in the optimal format for analysis and visualization.

### **4. LLM Integration**
- **Prompt Generation and Insight Creation**: The `LLMProcessor` dynamically generates prompts using the `promptgen` component. These prompts are tailored to the specific task (e.g., general analysis, summarization) and are sent to GPT-4 via the OpenAI API. The LLM returns insights, summaries, or other relevant information based on the prompt.

### **5. Visualization**
- **Data and Insights Visualization**: The `VisualizationAgent` creates visualizations of the processed data and the insights generated by the LLM. These visualizations include time series plots, correlation matrices, sentiment analysis charts, and more, depending on the data type.

### **6. Workflow Management**
- **Coordinated Workflow**: The `AgentManager` coordinates the entire workflow, ensuring that each agent (FileHandler, DataProcessingAgent, LLMProcessor, VisualizationAgent) performs its task in the correct sequence. The results are then passed back to the Streamlit UI for display.

### **7. Final Results Display (Streamlit UI)**
- **Interactive Display**: The final processed data, visualizations, and LLM-generated insights are presented to the user in the Streamlit interface. Users can further interact with the data, request additional insights, or upload new datasets for analysis.

# **b. data_processing_agent.py**
This module handles data processing tasks, including loading, cleaning, feature engineering, and transforming data.

````python
# agentlite_fraud_detection/data_processing_agent.py

import pandas as pd

class BaseDataProcessor:
    def process_data(self, data):
        raise NotImplementedError("This method should be overridden by subclasses")

    def clean_data(self, data):
        raise NotImplementedError("This method should be overridden by subclasses")

    def feature_engineering(self, data):
        raise NotImplementedError("This method should be overridden by subclasses")

    def transform_data(self, data):
        raise NotImplementedError("This method should be overridden by subclasses")

class DataProcessingAgent(BaseDataProcessor):
    def process_data(self, data):
        # Clean the data
        data = self.clean_data(data)
        # Perform feature engineering
        data = self.feature_engineering(data)
        # Transform the data
        data = self.transform_data(data)
        return data

    def clean_data(self, data):
        data.dropna(inplace=True)
        return data

    def feature_engineering(self, data):
        data['Daily_Return'] = data['Close'].pct_change()
        data['50_MA'] = data['Close'].rolling(window=50).mean()
        return data

    def transform_data(self, data):
        data['Normalized_Close'] = (data['Close'] - data['Close'].mean()) / data['Close'].std()
        return data
````

# **c. llm_processor.py**
This module integrates with an LLM like OpenAI’s GPT-4 to generate insights and perform NLP tasks

````python
# agentlite_fraud_detection/llm_processor.py

from openai import ChatCompletion
from agentlite_fraud_detection.promptgen import PromptGenerator  # Import the promptgen module

class LLMProcessor:
    def __init__(self, model="gpt-4", api_key="your-openai-api-key"):
        self.model = model
        self.api_key = api_key
        self.client = ChatCompletion(api_key=api_key)
        self.prompt_generator = PromptGenerator()  # Initialize promptgen

    def generate_insights(self, data, task="general_analysis"):
        # Use promptgen to generate a prompt based on the data and task
        prompt = self.prompt_generator.generate_prompt(data, task=task)

        response = self.client.create(
            model=self.model,
            messages=[{"role": "system", "content": "You are an expert in financial data analysis."},
                      {"role": "user", "content": prompt}]
        )
        return response['choices'][0]['message']['content']

    def perform_nlp_task(self, data, task):
        # Generate task-specific prompt using promptgen
        prompt = self.prompt_generator.generate_prompt(data, task=task)
        result = self.generate_insights(data, task=task)
        return result
````

# **e. file_handler.py**
This module handles file uploads, supporting both ZIP and CSV formats

````python
# agentlite_fraud_detection/file_handler.py

import os
import zipfile
import pandas as pd

class FileHandler:
    def __init__(self, upload_dir="uploaded_files"):
        self.upload_dir = upload_dir
        if not os.path.exists(upload_dir):
            os.makedirs(upload_dir)

    def handle_uploaded_file(self, uploaded_file):
        file_path = os.path.join(self.upload_dir, uploaded_file.name)

        with open(file_path, "wb") as f:
            f.write(uploaded_file.getbuffer())

        if zipfile.is_zipfile(file_path):
            return self._extract_zip(file_path)
        elif uploaded_file.name.endswith('.csv'):
            return pd.read_csv(file_path)
        else:
            raise ValueError("Unsupported file type. Please upload a ZIP or CSV file.")

    def _extract_zip(self, zip_path):
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(self.upload_dir)

        extracted_files = [f for f in os.listdir(self.upload_dir) if f.endswith('.csv')]
        if extracted_files:
            return pd.read_csv(os.path.join(self.upload_dir, extracted_files[0]))
        else:
            raise ValueError("No CSV files found in the ZIP archive.")
````

## **g.Implement or Update promptgen.py**

````python
# agentlite_fraud_detection/promptgen.py

class PromptGenerator:
    def __init__(self):
        pass

    def generate_prompt(self, data, task="general_analysis"):
        """
        Generate a prompt based on the provided data and task.
        """
        if task == "general_analysis":
            return f"Analyze the following data and provide insights: {data.head().to_string()}."
        elif task == "summarize":
            return f"Summarize the following financial data: {data.head().to_string()}."
        elif task == "fraud_detection":
            return f"Identify potential fraudulent activities in the following transactions: {data.head().to_string()}."
        elif task == "sentiment_analysis":
            return f"Analyze the sentiment of the following customer reviews: {data.head().to_string()}."
        else:
            return f"Perform {task} on the following data: {data.head().to_string()}."
````

## **1.Updated DataProcessingAgent**

````python
# agentlite_fraud_detection/data_processing_agent.py

import pandas as pd
from sklearn.preprocessing import StandardScaler

class DataProcessingAgent:
    def process_data(self, data):
        # Clean the data
        data = self.clean_data(data)
        # Perform feature engineering
        data = self.feature_engineering(data)
        # Transform the data
        data = self.transform_data(data)
        return data

    def clean_data(self, data):
        # Remove any null values
        data.dropna(inplace=True)
        return data

    def feature_engineering(self, data):
        if 'Amount' in data.columns:
            # Example for credit card fraud detection
            data['Scaled_Amount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
            # Time-based feature (assuming 'Time' column exists)
            data['Hour'] = data['Time'].apply(lambda x: (x // 3600) % 24)
        elif 'Revenue' in data.columns:
            # Example for financial data
            data['Profit_Margin'] = data['Net_Income'] / data['Revenue']
            data['ROE'] = data['Net_Income'] / data['Equity']
        return data

    def transform_data(self, data):
        if 'Amount' in data.columns:
            data['Normalized_Amount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
        return data
````

2. Update the LLMProcessor for Specific Insights

````python
# agentlite_fraud_detection/llm_processor.py

from openai import ChatCompletion
from agentlite_fraud_detection.promptgen import PromptGenerator

class LLMProcessor:
    def __init__(self, model="gpt-4", api_key="your-openai-api-key"):
        self.model = model
        self.api_key = api_key
        self.client = ChatCompletion(api_key=api_key)
        self.prompt_generator = PromptGenerator()

    def generate_insights(self, data, task="general_analysis"):
        prompt = self.prompt_generator.generate_prompt(data, task=task)
        response = self.client.create(
            model=self.model,
            messages=[{"role": "system", "content": "You are an expert in financial data analysis."},
                      {"role": "user", "content": prompt}]
        )
        return response['choices'][0]['message']['content']

    def perform_nlp_task(self, data, task):
        prompt = self.prompt_generator.generate_prompt(data, task=task)
        result = self.generate_insights(data, task=task)
        return result
````

4. Update the Visualization (VisualizationAgent)

````python
# agentlite_fraud_detection/visualization_agent.py

import matplotlib.pyplot as plt
import seaborn as sns

class VisualizationAgent:
    def __init__(self):
        pass

    def visualize_data(self, data):
        if 'Amount' in data.columns:
            plt.figure(figsize=(14, 7))
            sns.histplot(data['Scaled_Amount'], kde=True, bins=50)
            plt.title('Distribution of Scaled Transaction Amounts')
            plt.show()
        elif 'Revenue' in data.columns:
            plt.figure(figsize=(14, 7))
            sns.scatterplot(x='Revenue', y='Profit_Margin', data=data)
            plt.title('Revenue vs. Profit Margin')
            plt.show()

    def visualize_correlation_matrix(self, data):
        plt.figure(figsize=(12, 8))
        sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
        plt.title('Correlation Matrix')
        plt.show()
````
